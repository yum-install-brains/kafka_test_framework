# Results
- В целом, lz4 немного лучше (убираем snappy из дальнейших тестов)
- Для событий размером 512 байт размер батча 8192 байт слишком маленький (высокая latency)
- Для событий размером 1000 байт размер батча 8192 и 16384 байт слишком маленький (высокая latency)
- Вероятно, размер батча должен быть больше размера сообщения в 30 и более раз для хорошего батчевания и низкой latency (?)
- The size of record batches from the producer to the broker plays a role in latency. Smaller batches lead to more requests and queuing as well as higher latency.
- Батчи размером 8192 и 16384 байт в дальнейших тестах убираем
- Число партиций не сильно влияет на производительность при использовании 1 продюсера (тем не менее этот параметр оказывает влияние при большом числе продюсеров и консьюмеров – это вывод не из эксперимента, а general knowledge)
- Число партиций в дальнейшних тестах будет изменено. Тесты будут проведены на 9 и 27 партициях
- outgoingByteRate и byteRate сильно коррелируют, в дальнейших тестах оставим только byteRate (The average number of bytes sent per second for a topic)
- recordsSec коррелирует с recordSendRate, поэтому в дальнейших тестах оставляем только recordSendRate
